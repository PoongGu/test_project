```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins')
df
```

```python
y1 = df['species'] 
print(y1.unique()) # unique를 통해 y1의 값 분류
y2 = df['island']
print(y2.unique())
y3 = df['sex']
print(y3.unique())
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='species') # hue: 종 별로 색깔을 구분
```

```python
sns.pairplot(df, hue='species', palette='husl') # palette: 고급 시각화 효과
```

```python
sns.pairplot(df, hue='species', palette='husl', markers=['o','s', 'd']) # markers: 도형 시각화
```

```python
sns.pairplot(df, hue='species', palette='husl', markers=['o','s', 'd'], kind='reg') # kind: 분포도마다 회귀선 추가
```

```python
sns.pairplot(df, hue='species', palette='husl', markers=['o','s', 'D'], kind='reg', plot_kws={'line_kws' : {'color':'red'}}) # 회귀선 라인 색상 추가
```
